```python
# ESTRUTURA DE LOOP WHILE

Loop = 0

while Loop <= 10:
  print("Loop:", Loop)
  Loop+=1
```

```python
# USO DO -= (CONTAGEM REGRESSIVA)

contador = 10

while contador > 0:
    print("Contagem:", contador)
    contador -=1
```

```python
# ESTRUTURA DE LOOP WHILE

soma = 0
numero = 0

while numero != -1:
    numero = int(input("Digite um número:  "))
    if numero != -1:
        soma += numero
print("A soma dos números é:", soma)
```

```python
# ESTRUTURA DE LOOP WHILE

senha_correta = "1234"
tentativa = ""

while tentativa != senha_correta:
    tentativa = input("Digite sua senha: ")

print("Acesso permitido!")
```

```python
### ATIVIDADE ###
## Questão 1 – Contagem Regressiva
#Crie um programa em Python que utilize um laço while para imprimir os números de 10 até 0 (em ordem decrescente). Ao final, exiba a mensagem: "Fim da contagem!".

contador = 10

while contador > 0:
    print("Contagem:", contador)
    contador -=1
print("Fim da contagem")
```

```python
### ATIVIDADE ###
## Questão 2 – Apenas números pares
# Escreva um programa em Python que use um laço while para exibir todos os números pares de 0 até 20. Dica: um número é par quando numero % 2 == 0.

contador = 0

while contador <= 20:
    if contador % 2 == 0:
        print(contador)
    contador += 1
```

```python
###ATIVIDADE ###
## Questão 3
## Soma até -5 Crie um programa em Python que: Solicite ao usuário que digite vários números inteiros. O programa deve continuar pedindo números até que o usuário digite -5.
# Ao final, o programa deve mostrar a soma de todos os números digitados, sem incluir o -5 na soma. Utilize apenas duas variáveis no programa para resolver essa tarefa.

soma = 0
numero = 0

while numero != -5:
    numero = int(input("Digite um número:  "))
    if numero != -5:
        soma += numero
print("A soma total é:", soma)
```

```python
### ATIVIDADE ###
## Questão 4 – Código de Acesso ao Cofre
#Você foi contratado para programar o painel de acesso a um cofre digital. O cofre só pode ser aberto se o usuário digitar o código correto, que é "cofre2025". Crie um programa que: Peça ao usuário que digite o código do cofre.
#Continue pedindo até que o código correto seja digitado. Quando o código estiver correto, mostre a mensagem: "Cofre aberto com sucesso!"

codigo_correto = "cofre2025"
tentativa = ""

while tentativa != codigo_correto:
    tentativa = input("Digite seu codigo: ")

print("Cofre aberto com sucesso!")

```
